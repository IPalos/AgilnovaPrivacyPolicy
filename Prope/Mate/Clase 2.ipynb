**Ejemplo: Desintegración Radioactiva**

\begin{align}
\frac{dN}{dt} = -\alpha N \\[10pt]
N(t) = N(0)e^{-\alpha t} \\[10pt]
N(\tau) = \frac{N_0}{2} = N_0e^{-\alpha t} \\[10pt]
\therefore \tau = \frac{ln(2)}{\alpha}
\end{align}
**Vida media:** Tiempo al cual el número de atomos radioactivos decrece al 50%

```python
t = np.arange(0.0, 50000, 1)
a=1.23e-4
n0=1
v = n0*np.exp(-a*t)

fig, ax = plt.subplots(figsize=(8, 6), dpi=80)
ax.plot(t, v)

ax.set(xlabel='tiempo(años)', ylabel='porcentaje inicial',
       title='decaimiento radioactivo')
ax.grid()

plt.show()
```

**Ejemplo: Crecimiento de poblaciones (Ecuación Logística)**

\begin{align}
\frac{dx}{dt} &= r*x(t)-bx^2(t) \\[10pt]
\text{Condición Inicial: }x(0) &= x_0\\[10pt]
\text{Solución:}\\[10pt]
\frac{dx}{dt} &= rx(t)\left[ 1- \frac{b}{r} x(t) \right] \\[10pt]
\therefore \frac{dx}{rx \left[ 1- \frac{b}{r} x \right]} &= dt \\[10pt]
\int^x_{x_0} \frac{dx}{x \left[ 1- \frac{b}{r} x \right]} &= rt \\[10pt]
\int^x_{x_0} \left[\frac{1}{bx} +\frac{1}{1-bx} \right] dx &= rt \\[10pt]
\frac{1}{b}ln(x) -\frac{1}{b}ln(1-bx) &= \frac{r}{b}t \\[10pt]
\therefore x(t) &= \frac{K}{1+\frac{K-x_0}{x_0}e^{-rt}} \quad ; \quad K=\frac{r}{b}
\end{align}


```python
t = np.arange(0.0, 100, .01)
r=0.33
b=.00033
k=r/b
x0=100
x = k/(1+ ( ( (k-x0)/x0 )*np.exp(-r*t) ) )

fig, ax = plt.subplots(figsize=(8, 6), dpi=80)
ax.plot(t, x)

ax.set(xlabel='tiempo(años)', ylabel='porcentaje inicial',
       title='decaimiento radioactivo')
ax.grid()

plt.show()
```
